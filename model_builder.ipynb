```python
import pandas as pd
import os

orders_df = pd.read_csv("Data/olist_orders_dataset.csv")
customers_df = pd.read_csv("Data/olist_customers_dataset.csv")

order_item = pd.read_csv("Data/olist_order_items_dataset.csv")
product = pd.read_csv("Data/olist_products_dataset.csv")
```

```python
# Order X Customer Join 
order_customer_model = pd.merge(orders_df , customers_df ,on = 'customer_id' , how='inner')
```

```python


order_customer_model['order_purchase_timestamp'] = order_customer_model['order_purchase_timestamp'].astype('datetime64[ns]')
order_customer_model.info()


```

```python
try:
    os.mkdir('DataModel')
except:
    pass

order_customer_model.to_csv("DataModel/OrdersCustomers.csv",index=False)
```

```python
# Order X Customer Join 
order_customer_model = pd.merge(orders_df , customers_df[['customer_id','customer_unique_id']],on = 'customer_id' , how='left')

#Order X Customer X Order Itme 
order_customer_oitem = pd.merge(order_customer_model , order_item[['order_id','product_id','price','order_item_id'] ], on = 'order_id' , how='inner' )

# Order X Customer X Order Items X Products Dataset 
order_customer_product = pd.merge(order_customer_oitem , product[['product_id','product_category_name']] , on='product_id',how='inner')

agg_columns = {col : "first" for col in order_customer_product.columns}

agg_columns['price'] = "sum"
agg_columns['order_item_id'] = 'nunique'
order_customer_product = order_customer_product.groupby(['order_id','product_id'],as_index=False).agg(agg_columns)
```

```python
order_customer_product.to_csv("DataModel/OrderCustomerProduct.csv",index=False)
```
